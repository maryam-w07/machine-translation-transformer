```python
import os
import json
import unicodedata
from datasets import load_from_disk
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2CTCTokenizer, Wav2Vec2Processor

# === Paths ===
input_root   = "/content/cantonnese_batches"
output_root  = "/content/drive/MyDrive/tokenized_cantonnese"
vocab_path   = "vocab.json"
mapping_path = "mapping.json"

# === Load vocab and mapping ===
with open(vocab_path, "r") as f:
    vocab = json.load(f)
with open(mapping_path, "r") as f:
    mapping = json.load(f)

# === Prepare raw mapping keys (longest-first) ===
map_keys = sorted(mapping.keys(), key=len, reverse=True)

def raw_map(word, mapping, map_keys):
    i, L, out = 0, len(word), []
    while i < L:
        for key in map_keys:
            if word.startswith(key, i):
                out.append(mapping[key])
                i += len(key)
                break
        else:
            out.append(word[i])
            i += 1
    return "".join(out)

def segment_ipa(phoneme_str):
    segments = []
    current = ""
    for ch in phoneme_str:
        cat = unicodedata.category(ch)
        if cat.startswith("M") or cat in ("Lm", "Sk"):
            current += ch
        elif ch == " ":
            continue  # skip stray spaces
        else:
            if current:
                segments.append(current)
            current = ch
    if current:
        segments.append(current)
    return segments

def is_punct_or_empty(s):
    s = s.strip()
    if not s:
        return True
    return all(unicodedata.category(ch)[0] in ("P", "Z") for ch in s)

# === Setup Wav2Vec2 processor ===
tokenizer = Wav2Vec2CTCTokenizer(
    vocab_path, unk_token="[UNK]", pad_token="[PAD]", word_delimiter_token="|"
)
feature_extractor = Wav2Vec2FeatureExtractor(
    feature_size=1, sampling_rate=16000,
    padding_value=0.0, do_normalize=True,
    return_attention_mask=False
)
processor = Wav2Vec2Processor(
    feature_extractor=feature_extractor, tokenizer=tokenizer
)

# === Main sample processor ===
def prepare_sample(batch):
    batch["input_values"] = processor.feature_extractor(
        batch["audio_array"], sampling_rate=16000
    ).input_values[0]

    phoneme_words = []
    for raw in batch["phonemes"]:
        raw = raw.strip()
        if not raw:
            continue

        if raw == ",":
            phoneme_words.append("|")
            continue

        cleaned = raw_map(raw, mapping, map_keys)
        segments = segment_ipa(cleaned)
        segments = [s for s in segments if not is_punct_or_empty(s)]

        if segments:
            phoneme_words.append("".join(segments))

    # Filter empty and build full phoneme string with delimiters
    phoneme_string = "|".join(w for w in phoneme_words if w)

    batch["phonemes"] = phoneme_string
    batch["labels"] = processor.tokenizer(phoneme_string).input_ids
    return batch

# === Apply to all batches ===
batch_dirs = sorted(os.listdir(input_root))
for batch_name in batch_dirs:
    batch_path = os.path.join(input_root, batch_name)
    if not os.path.isdir(batch_path):
        continue

    print(f"Processing batch: {batch_name}")
    ds = load_from_disk(batch_path)
    processed = ds.map(prepare_sample)
    outp = os.path.join(output_root, batch_name)
    processed.save_to_disk(outp)
    print(f"✅ Saved to {outp}")
```

```python
import torch
from dataclasses import dataclass
from typing import Any, Dict, List, Optional, Union
from transformers import Wav2Vec2Processor

@dataclass
class DataCollatorCTCWithPadding:
    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True
    max_length: Optional[int] = None
    max_length_labels: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None
    pad_to_multiple_of_labels: Optional[int] = None

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        """
        Custom Data Collator for Phoneme-based CTC Training.
        Handles input audio and phoneme labels for batch processing.
        """
        # Debug: Verify that each feature has a valid audio array
        for idx, feature in enumerate(features):
            # Try to use "input_ids" first, else "input_values"
            input_val = feature.get("input_ids", feature.get("input_values"))
            if input_val is None:
                raise ValueError(f"Feature at index {idx} is missing both 'input_ids' and 'input_values': {feature}")
            # Optionally: check that the input array is not empty
            if hasattr(input_val, "__len__") and len(input_val) == 0:
                raise ValueError(f"Feature at index {idx} has an empty audio array: {feature}")

        # Prepare input features: use "input_values" field for audio data
        input_features = [
            {"input_values": feature.get("input_ids", feature.get("input_values"))}
            for feature in features
        ]

        # Debug: Verify that each feature has labels
        for idx, feature in enumerate(features):
            if "labels" not in feature:
                raise KeyError(f"Feature at index {idx} is missing the 'labels' key: {feature}")
        label_features = [{"input_ids": feature["labels"]} for feature in features]

        # Pad audio inputs using the processor
        batch = self.processor.pad(
            input_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="pt",
        )

        # Pad phoneme labels using the target processor
        with self.processor.as_target_processor():
            labels_batch = self.processor.pad(
                label_features,
                padding=self.padding,
                max_length=self.max_length_labels,
                pad_to_multiple_of=self.pad_to_multiple_of_labels,
                return_tensors="pt",
            )

        # Replace padding tokens in labels with -100 so they're ignored during loss computation.
        # NOTE: Ensure that your custom tokenizer has the correct pad_token_id for your 42-phoneme vocabulary.
        labels = labels_batch["input_ids"]
        labels[labels == self.processor.tokenizer.pad_token_id] = -100

        batch["labels"] = labels
        return batch
```

```python
import torch
import numpy as np


def compute_metrics(pred):
    pred_logits = pred.predictions
    pred_ids = np.argmax(pred_logits, axis=-1)

    # Replace -100 with pad_token_id for decoding
    labels = np.copy(pred.label_ids)
    labels[labels == -100] = processor.tokenizer.pad_token_id

    # Decode predictions and labels
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = processor.batch_decode(labels, skip_special_tokens=True)

    # Compute WER
    wer_score = wer_metric.compute(predictions=pred_str, references=label_str)
    return {"wer": wer_score}

def map_to_result(batch, model):
    # Corrected indentation here
    with torch.no_grad():
        input_values = torch.tensor(batch["input_values"])

        # Ensure input_values is 2D: (batch_size, sequence_length)
        if input_values.ndim == 1:
            input_values = input_values.unsqueeze(0)

        input_values = input_values.to(model.device)

        logits = model(input_values).logits

    pred_ids = torch.argmax(logits, dim=-1)

    # Decode all predictions in batch
    pred_strs = processor.batch_decode(pred_ids, skip_special_tokens=True)

    # Decode all targets
    label_ids = batch["labels"]
    if isinstance(label_ids[0], list):  # batched case
        target_strs = processor.batch_decode(label_ids, group_tokens=False)
    else:  # single sample
        label_ids = [label_ids]
        target_strs = [processor.decode(label_ids[0], group_tokens=False)]

    # Attach results to batch
    batch["pred_ids"] = [ids.tolist() for ids in pred_ids]
    batch["pred_str"] = pred_strs
    batch["sentence"] = target_strs

    # Return the modified batch
    return batch
```

```python
import torch
import torchaudio
import torchaudio.functional as F
from transformers import AutoProcessor, AutoModelForCTC, Wav2Vec2Processor
from transformers import Wav2Vec2ForCTC, Wav2Vec2FeatureExtractor, Wav2Vec2CTCTokenizer
from datasets import load_dataset, load_from_disk
from itertools import groupby
from dataclasses import dataclass
import json
import os
import argparse
import datetime
from pathlib import Path
import sys
import gradio as gr
import numpy as np
import nltk
import re
# from g2p_en import G2p
# import phonemizer
import logging
# from phonemizer.backend import EspeakBackend

# logging.getLogger('phonemizer').setLevel(logging.ERROR)

# ru_phonemizer_backend = EspeakBackend('ru')
# en_phonemizer_backend = EspeakBackend('en-us')

def decode_phonemes(
    ids: torch.Tensor, processor: Wav2Vec2Processor, ignore_stress: bool = False
) -> str:
    """CTC-like decoding. First removes consecutive duplicates, then removes special tokens."""
    # removes consecutive duplicates
    ids = [id_ for id_, _ in groupby(ids)]

    # special_token_ids = processor.tokenizer.all_special_ids + [
    #     processor.tokenizer.word_delimiter_token_id
    # ]
    special_token_ids = processor.tokenizer.all_special_ids

    # converts id to token, skipping special tokens
    phonemes = [processor.decode(id_) for id_ in ids if id_ not in special_token_ids]

    # joins phonemes
    prediction = " ".join(phonemes)

    # whether to ignore IPA stress marks
    if ignore_stress == True:
        prediction = prediction.replace("ˈ", "").replace("ˌ", "")

    return prediction

def get_trellis(emission, tokens, blank_id=0):
    num_frame = emission.size(0)
    num_tokens = len(tokens)

    trellis = torch.zeros((num_frame, num_tokens))
    trellis[1:, 0] = torch.cumsum(emission[1:, blank_id], 0)
    trellis[0, 1:] = -float("inf")
    trellis[-num_tokens + 1 :, 0] = float("inf")

    for t in range(num_frame - 1):
        trellis[t + 1, 1:] = torch.maximum(
            # Score for staying at the same token
            trellis[t, 1:] + emission[t, blank_id],
            # Score for changing to the next token
            trellis[t, :-1] + emission[t, tokens[1:]],
        )
    return trellis

@dataclass
class Point:
    token_index: int
    time_index: int
    score: float

def backtrack(trellis, emission, tokens, blank_id=0):
    t, j = trellis.size(0) - 1, trellis.size(1) - 1

    path = [Point(j, t, emission[t, blank_id].exp().item())]
    while j > 0:
        # Should not happen but just in case
        assert t > 0

        # 1. Figure out if the current position was stay or change
        # Frame-wise score of stay vs change
        p_stay = emission[t - 1, blank_id]
        p_change = emission[t - 1, tokens[j]]

        # Context-aware score for stay vs change
        stayed = trellis[t - 1, j] + p_stay
        changed = trellis[t - 1, j - 1] + p_change

        # Update position
        t -= 1
        if changed > stayed:
            j -= 1

        # Store the path with frame-wise probability.
        prob = (p_change if changed > stayed else p_stay).exp().item()
        path.append(Point(j, t, prob))

    # Now j == 0, which means, it reached the SoS.
    # Fill up the rest for the sake of visualization
    while t > 0:
        prob = emission[t - 1, blank_id].exp().item()
        path.append(Point(j, t - 1, prob))
        t -= 1

    return path[::-1]

@dataclass
class Segment:
    label: str
    start: int
    end: int
    score: float

    def __repr__(self):
        return f"{self.label}\t({self.score:4.2f}): [{self.start:5d}, {self.end:5d})"

    @property
    def length(self):
        return self.end - self.start

def merge_repeats(path, transcript):
    i1, i2 = 0, 0
    segments = []
    while i1 < len(path):
        while i2 < len(path) and path[i1].token_index == path[i2].token_index:
            i2 += 1
        score = sum(path[k].score for k in range(i1, i2)) / (i2 - i1)
        segments.append(
            Segment(
                transcript[path[i1].token_index],
                path[i1].time_index,
                path[i2 - 1].time_index + 1,
                score,
            )
        )
        i1 = i2
    return segments

def get_phoneme2id_dict(phoneme2id_file):
    with open(phoneme2id_file, 'r') as file:
        vocab_dict = json.load(file)

    return vocab_dict

def get_phoneme2viseme_dict(phoneme2viseme_file):
    with open(phoneme2viseme_file, 'r') as file:
        p2v_dict = json.load(file)

    return p2v_dict

def remove_special_characters(text):
    chars_to_ignore_regex = '[\,\?\.\!\-\;\:\"]'
    normalized_text = re.sub(chars_to_ignore_regex, '', text).lower()
    return normalized_text

def forced_alignment_streaming_overlap_vad(input_audio_file, input_text_file, model, g2p, vocab_dict, tokens_to_phonemes, phoneme2viseme_map, processor, device, vad_model, vad_utils, stats_flag=False):

    (get_speech_timestamps, _, _, _, _) = vad_utils

    overlap_ratio =  1/4
    sil_thresh = 0.1
    sample_rate = 16000
    chunk_len = 16000 #16000 #16000
    overlap_len = int(sample_rate*overlap_ratio)
    step_len = chunk_len - overlap_len

    # audio, osr = torchaudio.load(input_audio_file)
    # if osr != 16000:
    #     audio = F.resample(audio, osr, sample_rate)
    #     print(f'Original audio rate: {osr}, resampled at:{sample_rate}')

    audio = input_audio_file
    # audio = audio.to(device)

    viseme_json_out = []
    last_chunk_end_duration = 0

    for chunk_num, start in enumerate(range(0, audio.shape[1], step_len)):

        chunk_viseme_out = []

        # sttime = time.time()
        chunk = audio[:, start:start+chunk_len]

        speech_timestamps = get_speech_timestamps(chunk, vad_model, threshold=0.7, sampling_rate=sample_rate, min_speech_duration_ms=50, return_seconds=True)

        # Complete silence in the chunk
        if len(speech_timestamps) == 0:
            if chunk_num == 0:
                data = {'duration': 0.0, 'visemeId': phoneme2viseme_map['|']}
                viseme_json_out.append(data)
            else:
                if viseme_json_out[-1]['visemeId'] == phoneme2viseme_map['|']:
                    continue
                else:
                    data = {'duration': round(last_chunk_end_duration + overlap_ratio*1000, 3), 'visemeId': phoneme2viseme_map['|']}
                    viseme_json_out.append(data)
            last_chunk_end_duration += 1000.0 - overlap_ratio*1000
            continue

        # load each chunk to device
        chunk = chunk.to(device)

        # Perform inference
        logits = (model(chunk).logits).detach().cpu()
        emission = logits.squeeze(0)

        # Decode the logits to get the predicted text
        pred_ids = torch.argmax(logits, dim=-1)
        pred_phonemes = processor.batch_decode(pred_ids[0])

        # group phonemes
        group_phonemes_list = []
        # Frist group phonemes, '[PAD]' and ''
        for p in pred_phonemes:
            if len(group_phonemes_list) == 0:
                group_phonemes_list.append(p)
            elif group_phonemes_list[-1] == p:
                continue
            else:
                group_phonemes_list.append(p)

        # remove '[PAD]'
        group_phonemes_list = [p for p in group_phonemes_list if p!='[PAD]']

        # Replace '' with '|', 'H' with 'HH', 'A' with 'AA'
        for pnum, p in enumerate(group_phonemes_list):
            if p == '':
                group_phonemes_list[pnum] = '|'
        #     elif p == 'H':
        #         group_phonemes_list[pnum] = 'HH'
        #     elif p == 'A':
        #         group_phonemes_list[pnum] = 'AA'
        # import pdb; pdb.set_trace()
        tokens = []
        for p in group_phonemes_list:
            if p == ' ':
                tokens.append(vocab_dict['|'])
            else:
                tokens.append(vocab_dict[p])

        if len(tokens) == 0:
            continue

        # if start == 0:
        #     if tokens[0] != vocab_dict['|']: tokens.insert(0, vocab_dict['|'])
        # if tokens[0] != vocab_dict['|']: tokens.insert(0, vocab_dict['|'])

        # get trellis, path and segments
        trellis = get_trellis(emission, tokens, blank_id=vocab_dict['[PAD]'])
        path = backtrack(trellis, emission, tokens, blank_id=vocab_dict['[PAD]'])
        transcript = [tokens_to_phonemes[t] for t in tokens]
        segments = merge_repeats(path, transcript)

        # ----- VAD + segments combination to get visemes. ----- #
        ratio = chunk.size(1) / sample_rate / trellis.size(0)

        # generate visemes
        for phoneme_num, phoneme in enumerate(segments):
            x0 = ratio * phoneme.start
            x1 = ratio * phoneme.end

            # First 1-sec chunk
            if last_chunk_end_duration == 0:
                if phoneme_num ==0:
                    data = {'duration': 0.0, 'visemeId': phoneme2viseme_map['|']}
                    viseme_json_out.append(data)
                    data = {'duration': round(speech_timestamps[0]['start']*1000 + 10, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                    viseme_json_out.append(data)
                else:
                    if phoneme.label == '|' and ((x1-x0) < sil_thresh):
                        continue

                    data = {'duration': round(x0*1000, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                    if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map[phoneme.label]:
                        viseme_json_out.append(data)

            # 2nd and onward chunks
            else:
                if x1 < overlap_ratio:
                    continue
                else:
                    if x0 < overlap_ratio:
                        if x1-x0 < sil_thresh or viseme_json_out[-1]['visemeId'] == phoneme2viseme_map[phoneme.label]:
                            continue

                        if speech_timestamps[0]['start'] > sil_thresh:
                            data = {'duration': round(last_chunk_end_duration + overlap_ratio*1000, 3), 'visemeId': phoneme2viseme_map['|']}
                            if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map['|']:
                                viseme_json_out.append(data)
                            data = {'duration': round(last_chunk_end_duration + x1*1000, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                            viseme_json_out.append(data)
                        else:
                            data = {'duration': round(last_chunk_end_duration + overlap_ratio*1000, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                            if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map[phoneme.label]:
                                viseme_json_out.append(data)
                    else:
                        if x0 < speech_timestamps[0]['start']:
                            if (speech_timestamps[0]['start']-x0) > 0.1:
                                data = {'duration': round(last_chunk_end_duration + x0*1000, 3), 'visemeId': phoneme2viseme_map['|']}
                                if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map['|']:
                                    viseme_json_out.append(data)
                                data = {'duration': round(last_chunk_end_duration + speech_timestamps[0]['start']*1000, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                                if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map[phoneme.label]:
                                    viseme_json_out.append(data)
                            else:
                                data = {'duration': round(last_chunk_end_duration + speech_timestamps[0]['start']*1000, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                                if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map[phoneme.label]:
                                    viseme_json_out.append(data)
                        else:
                            # skipping short silence segment
                            if phoneme.label == '|' and ((x1-x0) < sil_thresh):
                                continue

                            data = {'duration': round(last_chunk_end_duration + x0*1000, 3), 'visemeId': phoneme2viseme_map[phoneme.label]}
                            if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map[phoneme.label]:
                                viseme_json_out.append(data)

        # Add last silence viseme if there is no speech detected.
        if speech_timestamps[-1]['end'] != 1:
            data = {'duration': round(last_chunk_end_duration + speech_timestamps[-1]['end']*1000, 3), 'visemeId': phoneme2viseme_map['|']}
            if viseme_json_out[-1]['visemeId'] != phoneme2viseme_map['|']:
                viseme_json_out.append(data)

        last_chunk_end_duration += round(x1*1000.0,3) - overlap_ratio*1000


    viseme_out_file = 'output_gr_viseme.json'

    # save viseme to json file
    # outfile_name = os.path.basename(input_audio_file)[:-4]
    json_object = json.dumps(viseme_json_out, indent=4)
    with open(viseme_out_file, "w") as outfile:
        outfile.write(json_object)

    return viseme_out_file, f'File available to download: {viseme_out_file}'


def forced_alignment_streaming_overlap(input_audio_file, input_text_file, model, g2p, vocab_dict, tokens_to_phonemes, phoneme2viseme_map, processor, device):

    overlap_ratio =  1/4
    sil_thresh = 0.1
    sample_rate = 16000
    chunk_len = 16000 #0.5s
    overlap_len = int(chunk_len*overlap_ratio)
    step_len = chunk_len - overlap_len

    # audio, osr = torchaudio.load(input_audio_file)
    # if osr != 16000:
    #     audio = F.resample(audio, osr, sample_rate)
    #     print(f'Original audio rate: {osr}, resampled at:{sample_rate}')

    audio = input_audio_file

    audio = audio.to(device)

    # while True:
    pred_str_complete = ''
    viseme_json_out = []
    last_chunk_end_duration = 0

    for chunk_num, start in enumerate(range(0, audio.shape[1], step_len)):
        # sttime = time.time()
        chunk = audio[:, start:start+chunk_len]

        # Perform inference
        logits = (model(chunk).logits).detach().cpu()
        emission = logits.squeeze(0)

        # Decode the logits to get the predicted text
        pred_ids = torch.argmax(logits, dim=-1)
        pred_phonemes = processor.batch_decode(pred_ids[0])

        # group phonemes
        group_phonemes_list = []
        # Frist group phonemes, '[PAD]' and ''
        for p in pred_phonemes:
            if len(group_phonemes_list) == 0:
                group_phonemes_list.append(p)
            elif group_phonemes_list[-1] == p:
                continue
            else:
                group_phonemes_list.append(p)

        # remove '[PAD]'
        group_phonemes_list = [p for p in group_phonemes_list if p!='[PAD]']
        # Replace '' with '|', 'H' with 'HH', 'A' with 'AA'
        for pnum, p in enumerate(group_phonemes_list):
            if p == '':
                group_phonemes_list[pnum] = '|'
            elif p == 'H':
                group_phonemes_list[pnum] = 'HH'
            elif p == 'A':
                group_phonemes_list[pnum] = 'AA'

        tokens = []
        for p in group_phonemes_list:
            if p == ' ':
                tokens.append(vocab_dict['|'])
            else:
                tokens.append(vocab_dict[p])

        if len(tokens) == 0:
            continue

        # if start == 0:
        #     if tokens[0] != vocab_dict['|']: tokens.insert(0, vocab_dict['|'])
        if tokens[0] != vocab_dict['|']: tokens.insert(0, vocab_dict['|'])

        # get trellis, path and segments
        trellis = get_trellis(emission, tokens, blank_id=vocab_dict['[PAD]'])
        path = backtrack(trellis, emission, tokens, blank_id=vocab_dict['[PAD]'])
        transcript = [tokens_to_phonemes[t] for t in tokens]
        segments = merge_repeats(path, transcript)

        ratio = chunk.size(1) / sample_rate / trellis.size(0)
        phoneme_time_segments = []
        # generate visemes
        for phoneme in segments:
            x0 = ratio * phoneme.start
            x1 = ratio * phoneme.end
            ####### skipping short silence segments #######
            if phoneme.label == '|' and ((x1-x0) < sil_thresh):
                continue
            ###############################################
            phoneme_time_segments.append([phoneme.label, x0, x1, phoneme.score])
            # data = {'duration': last_chunk_end_duration + round(x0*1000.0,3), 'visemeId': phoneme2viseme_map[phoneme.label]}
            data = {'duration': round(last_chunk_end_duration + x0*1000.0,3), 'visemeId': phoneme2viseme_map[phoneme.label]}

            if len(viseme_json_out) > 0:
                if (last_chunk_end_duration + round(x0*1000.0,3)) >= viseme_json_out[-1]['duration']:
                    viseme_json_out.append(data)
            else:
                viseme_json_out.append(data)

        last_chunk_end_duration += round(x1*1000.0,3) - overlap_ratio*1000

    viseme_out_file = 'output_gr_viseme.json'

    # save viseme to json file
    # outfile_name = os.path.basename(input_audio_file)[:-4]
    json_object = json.dumps(viseme_json_out, indent=4)
    with open(viseme_out_file, "w") as outfile:
        outfile.write(json_object)

    return viseme_out_file, f'File available to download: {viseme_out_file}'


def forced_alignment_given_audio_text(input_audio_file, input_text_file, model, g2p, vocab_dict, tokens_to_phonemes, phoneme2viseme_map, device):

    sil_thresh = 0.1
    sample_rate = 16000
    # audio, sr = torchaudio.load(input_audio_file)
    # if sr != 16000:
    #     audio = F.resample(audio, sr, sample_rate)

    audio = input_audio_file
    # load audio to device (cuda/cpu)
    audio = audio.to(device)

    # read text from input text file.
    # with open(input_text_file, 'r') as fo:
    #     input_text = fo.readline()
    input_text = input_text_file

    # convert phonemes to tokens (ids)
    # phonemes_list = g2p(input_text)[:-1]
    input_text = remove_special_characters(input_text)
    phonemes_list = g2p(input_text)
    phonemes_list = [p.replace('0','').replace('1','').replace('2','') for p in phonemes_list]

    tokens = []
    for p in phonemes_list:
        if p == ' ':
            tokens.append(vocab_dict['|'])
        else:
            tokens.append(vocab_dict[p])

    # compute model output logits
    logits = model(audio).logits
    emission = (logits.squeeze(0)).detach().to('cpu')

    # Append space tokens at the start and end.
    if tokens[0] != vocab_dict['|']: tokens.insert(0, vocab_dict['|'])
    if tokens[-1]!= vocab_dict['|']: tokens.append(vocab_dict['|'])

    # get trellis, path and segments
    trellis = get_trellis(emission, tokens, blank_id=vocab_dict['[PAD]'])
    path = backtrack(trellis, emission, tokens, blank_id=vocab_dict['[PAD]'])
    transcript = [tokens_to_phonemes[t] for t in tokens]
    segments = merge_repeats(path, transcript)

    audio = audio.squeeze(0)
    ratio = audio.size(0) / sample_rate / trellis.size(0)

    # Align segments with the audio time stamps
    phoneme_time_segments = []
    for phoneme in segments:
        x0 = ratio * phoneme.start
        x1 = ratio * phoneme.end
        phoneme_time_segments.append([phoneme.label, x0, x1])

    viseme_out_file = 'output_gr_viseme.json'

    viseme_json_out = []
    for p in phoneme_time_segments:
        ####### skipping short silence segments #######
        if p[0] == '|' and (p[2]-p[1] < sil_thresh):
            continue
        ###############################################
        data = {'duration': round(p[1]*1000.0,3), 'visemeId': phoneme2viseme_map[p[0]]}
        viseme_json_out.append(data)

    # appending silence at the end.
    data = {'duration': round(p[2]*1000.0,3), 'visemeId': phoneme2viseme_map['|']}
    viseme_json_out.append(data)

    # save viseme to json file
    json_object = json.dumps(viseme_json_out, indent=4)
    with open(viseme_out_file, "w") as outfile:
        outfile.write(json_object)

    # print(f'Output viseme (.json) file saved at: {viseme_out_file}')

    return viseme_out_file, f'File available to download: {viseme_out_file}'

if __name__ == "__main__":

    nltk.download('averaged_perceptron_tagger_eng')

    class FA_object:
        def __init__(self):
            self.phoneme2viseme_file = '/content/phoneme2viseme.json'
            self.phoneme2id_file = '/content/vocab.json'
            self.model_path = "/content/drive/MyDrive/wav2vec2_training/checkpoint-5500"
            #self.tokenizer_path = "cluster1_tokenizer"
            self.device = "cuda" if torch.cuda.is_available() else "cpu"
            self.output_audio_viseme_dir = 'output_audio_viseme'
            # self.dataset_path = 'dataset_audio_phoneme_2'

            # get phoneme2viseme mapping from text file.
            self.phoneme2viseme_map = get_phoneme2viseme_dict(self.phoneme2viseme_file)
            # Initialize the tokenizer from your vocab file for inference
            self.tokenizer = Wav2Vec2CTCTokenizer(
                self.phoneme2id_file,
                unk_token="[UNK]",
                pad_token="[PAD]",
                word_delimiter_token="|"
                )

            # import g2p model
            # self.g2p = GraphemeToPhoneme.from_hparams("speechbrain/soundchoice-g2p",
            #                                     savedir="pretrained_models/soundchoice-g2p",
            #                                     run_opts={"device":self.device}
            #                                     )

            self.g2p = None #{'ru': ru_phonemizer_backend, 'en-us': en_phonemizer_backend}

            self.vad_model, self.vad_utils = torch.hub.load(repo_or_dir='snakers4/silero-vad', model='silero_vad')

            # load model
            self.model = Wav2Vec2ForCTC.from_pretrained(self.model_path)
            self.model.to(self.device)

            # load processor
            #self.processor = Wav2Vec2Processor.from_pretrained(self.tokenizer_path)
            self.feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/wav2vec2-base-960h")

           # Combine into processor
            # Corrected Indentation here
            self.processor = Wav2Vec2Processor(feature_extractor=self.feature_extractor, tokenizer=self.tokenizer)


            self.vocab_dict = get_phoneme2id_dict(self.phoneme2id_file)
            self.tokens_to_phonemes = {}
            self.tokens_to_phonemes = dict([(value, key) for key, value in self.vocab_dict.items()])

        def process(self, input_audio_file, input_text_file, run_type='non-streaming'):

            if run_type == 'non-streaming':
                return forced_alignment_given_audio_text(
                    input_audio_file=input_audio_file,
                    input_text_file=input_text_file,
                    model=self.model,
                    g2p=self.g2p,
                    vocab_dict=self.vocab_dict,
                    tokens_to_phonemes=self.tokens_to_phonemes,
                    phoneme2viseme_map=self.phoneme2viseme_map,
                    device=self.device
                    )
            elif run_type == 'streaming':
                return forced_alignment_streaming_overlap(
                    input_audio_file=input_audio_file,
                    input_text_file=input_text_file,
                    model=self.model,
                    g2p=None,
                    vocab_dict=self.vocab_dict,
                    tokens_to_phonemes=self.tokens_to_phonemes,
                    phoneme2viseme_map=self.phoneme2viseme_map,
                    processor=self.processor,
                    device=self.device
                    )
            elif run_type == 'streaming_vad':
                return forced_alignment_streaming_overlap_vad(
                    input_audio_file=input_audio_file,
                    input_text_file=input_text_file,
                    model=self.model,
                    g2p=self.g2p,
                    vocab_dict=self.vocab_dict,
                    tokens_to_phonemes=self.tokens_to_phonemes,
                    phoneme2viseme_map=self.phoneme2viseme_map,
                    processor=self.processor,
                    device=self.device,
                    vad_model=self.vad_model,
                    vad_utils=self.vad_utils,
                    )

            else:
                print('Type error...')
                import sys; sys.exit()

    # Create an instance of your object
    fa_process_obj = FA_object()

    def wrapper_function(input_audio, input_text, streaming):

        sr, y = input_audio
        y = np.array(y, dtype=np.float32)
        y /= np.max(np.abs(y))

        if sr != 16000:
            audio = F.resample(torch.tensor(y).reshape(1,-1), sr, 16000)
        else:
            audio = torch.tensor(y).reshape(1,-1)

        return fa_process_obj.process(audio, input_text, streaming)


with gr.Blocks() as demo:

    gr.Markdown("# Provide audio file and transcirpt to download viseme (.json) file.")

    with gr.Row():

        with gr.Column(scale=1):
            # Input components
            audio = gr.Audio(label='input audio', sources="upload")
            # radio_button = gr.Radio(choices=['non-streaming', 'streaming', 'streaming_vad'], value='non-streaming', label='Model')
            radio_button = gr.Radio(choices=['streaming_vad'], value='streaming_vad', label='Model')
            # input textbox for transcript
            input_transcript = gr.Textbox(label='Input Transcript')

            # Submit button
            submit_button = gr.Button("Submit")

        with gr.Column(scale=1):
            # Output component
            notification_button = gr.Label(label='Output Notification')
            # Download button
            download_button = gr.DownloadButton(label='Download')

    # Define button click action
    submit_button.click(
        fn=wrapper_function,       # Function to call on submit
        inputs=[audio, input_transcript, radio_button], # Inputs from the form
        outputs=[download_button, notification_button]  # Output to display
    )

    demo.css = """
    .button-on {
        background-color: green;
        color: white;
    }
    .button-off {
        background-color: grey;
        color: white;
    }
    #togg_button {
        border: none;
        cursor: pointer;
    }
    """

demo.launch(share=True)
```
