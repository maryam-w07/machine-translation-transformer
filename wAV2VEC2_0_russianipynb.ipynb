```python
import torch
import torchaudio

TARGET_RATE = 16000

def resample_audio_sample(sample):
    """
    Resamples the audio in the sample to TARGET_RATE if necessary.
    Expects sample["audio"] to have keys "array" (waveform as NumPy array) and "sampling_rate".
    """
    waveform = sample["audio"]["array"]
    sr = sample["audio"]["sampling_rate"]

    # Convert waveform to torch.Tensor and cast to float32.
    if not isinstance(waveform, torch.Tensor):
        waveform = torch.tensor(waveform)
    waveform = waveform.float()  # waveform is float32

    if sr != TARGET_RATE:
        resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=TARGET_RATE)
        waveform = resampler(waveform)
        sample["audio"]["array"] = waveform.numpy()  # update with new array as numpy array
        sample["audio"]["sampling_rate"] = TARGET_RATE
    return sample


for split in ["train", "validation", "test"]:
    small_dataset[split] = [resample_audio_sample(sample) for sample in small_dataset[split]]


from IPython.display import Audio, display

for split in ["train"]:
    print(f"--- {split.upper()} Examples ---")
    for i, sample in enumerate(small_dataset[split]):
        if i >= 3:
            break
        audio_array = sample["audio"]["array"]
        sampling_rate = sample["audio"]["sampling_rate"]
        print("New sampling rate:", sampling_rate)
        display(Audio(audio_array, rate=sampling_rate))
```

```python
def apply_mapping(phoneme_str, mapping):
    """
    Apply the mapping to a phonemized string.
    The string is assumed to have words separated by "|" and phonemes within words separated by spaces.
    """
    # Split the string into words using the delimiter "|"
    words = phoneme_str.split("|")
    updated_words = []

    for word in words:
        # Remove extra whitespace and split the word into tokens (phonemes)
        tokens = word.strip().split()
        # Replace each token if it's in the mapping, otherwise leave it unchanged.
        updated_tokens = [mapping.get(token, token) for token in tokens]
        # Join the tokens back together with a single space.
        updated_words.append(" ".join(updated_tokens))

    # Reconstruct the phonemized string using " | " as the word delimiter.
    return " | ".join(updated_words)

# Apply the mapping to all splits in your phonemized dataset.
for split in ["train", "test", "validation"]:
    for sample in phonemized_dataset[split]:
        original = sample["phonemes"]
        sample["phonemes"] = apply_mapping(original, phoneme_mapping)

# Verification: Print the updated phonemes from a sample in each split.
print("Train sample phonemes:", phonemized_dataset["train"][0]["phonemes"])
print("Validation sample phonemes:", phonemized_dataset["validation"][0]["phonemes"])
print("Test sample phonemes:", phonemized_dataset["test"][0]["phonemes"])
```

```python
def post_process(phoneme_line: str) -> str:
    """
    Given a phoneme string where individual phonemes are separated by spaces and words
    are separated by the delimiter '|', this function:
      - Splits the line on '|'
      - For each word, strips extra spaces and joins its phonemes (removing intra-word spaces)
      - Joins the processed words with a single space
    """
    # Split the line on the word delimiter '|'
    words = phoneme_line.split('|')

    processed_words = []
    for word in words:
        # Strip leading/trailing whitespace from the word and split by whitespace to get phonemes
        phonemes = word.strip().split()
        # Join phonemes to form the complete word (i.e. remove spaces between phonemes)
        processed_word = ''.join(phonemes)
        # Only add non-empty words
        if processed_word:
            processed_words.append(processed_word)

    # Join the words with a space so that words are separated
    return ' '.join(processed_words)



# Process the entire dataset in place
for split in ["train", "validation", "test"]:
    for sample in phonemized_dataset[split]:
        # Update the phonemes field with the post-processed version
        sample["phonemes"] = post_process(sample["phonemes"])

# Print the first 10 samples from each split to verify the changes
for split in ["train", "validation", "test"]:
    print(f"\n🔹 {split.upper()} Samples After Processing:")
    for i, sample in enumerate(phonemized_dataset[split][:10]):
        print(f"{sample['sentence']} -> {sample['phonemes']}")

```

```python
def post_process(phoneme_line: str) -> str:
    # Split the line by the delimiter '|' to separate words
    words = phoneme_line.split('|')

    # For each word, remove extra spaces and join phonemes together
    processed_words = []
    for word in words:
        # Remove leading/trailing spaces and split by whitespace
        phonemes = word.strip().split()
        # Join phonemes without spaces to form the word
        processed_word = ''.join(phonemes)
        if processed_word:  # Only add non-empty words
            processed_words.append(processed_word)

    # Join the processed words with a space
    return ' '.join(processed_words)

# Sample phoneme strings from your dataset:
train_sample = "p ʌ ɛ t ʌ m u | n  m | nʲ ɛ | x v a t  j i t | ɪ n s t r u mʲ ɛ n t ʌ v | d lʲ  | a dʲ i k v  t n ʌ v ʌ | rʲ i a gʲ i r ʌ v ʌ nʲ i j a | n ə | ɛ tʲ ɪ | v ɨ z ʌ v ɨ |"
validation_sample = "k ɔ lʲ ʌ | n u | k  k | ʐ ɛ | t v ʌ j a | r a b ɔ t a | p l ɔ x ʌ |"
test_sample = "m a ʂ ʂ t  b ɨ | fʲ i n ʌ n s ʌ v ʌ ɛ k ʌ n ɔ mʲ i t͡ɕ i s k ʌ v ʌ | k rʲ i zʲ i s a | i | tʲ ɛ m p ɨ | j ɪ v ɔ | r ʌ s p r ʌ s t r a nʲ ɛ nʲ i j a | z a s t  lʲ ɪ | s  m ɨ x | ɔ p ɨ t n ɨ x | s pʲ i t͡s ɨ a lʲ i s t ʌ f | mʲ i r a | v r a s p l"

# Process each sample:
processed_train = post_process(train_sample)
processed_validation = post_process(validation_sample)
processed_test = post_process(test_sample)

# Print the processed outputs:
print("Processed Train Sample:")
print(processed_train)
print("\nProcessed Validation Sample:")
print(processed_validation)
print("\nProcessed Test Sample:")
print(processed_test)
```

```python
import torch
from dataclasses import dataclass
from typing import Any, Dict, List, Optional, Union
from transformers import Wav2Vec2Processor

@dataclass
class DataCollatorCTCWithPadding:
    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True
    max_length: Optional[int] = None
    max_length_labels: Optional[int] = None
    pad_to_multiple_of: Optional[int] = None
    pad_to_multiple_of_labels: Optional[int] = None

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        """
        Custom Data Collator for Phoneme-based CTC Training.
        Handles input audio and phoneme labels for batch processing.
        """
        # Debug: Verify that each feature has a valid audio array
        for idx, feature in enumerate(features):
            # Try to use "input_ids" first, else "input_values"
            input_val = feature.get("input_ids", feature.get("input_values"))
            if input_val is None:
                raise ValueError(f"Feature at index {idx} is missing both 'input_ids' and 'input_values': {feature}")
            # Optionally: check that the input array is not empty
            if hasattr(input_val, "__len__") and len(input_val) == 0:
                raise ValueError(f"Feature at index {idx} has an empty audio array: {feature}")

        # Prepare input features: use "input_values" field for audio data
        input_features = [
            {"input_values": feature.get("input_ids", feature.get("input_values"))}
            for feature in features
        ]

        # Debug: Verify that each feature has labels
        for idx, feature in enumerate(features):
            if "labels" not in feature:
                raise KeyError(f"Feature at index {idx} is missing the 'labels' key: {feature}")
        label_features = [{"input_ids": feature["labels"]} for feature in features]

        # Pad audio inputs using the processor
        batch = self.processor.pad(
            input_features,
            padding=self.padding,
            max_length=self.max_length,
            pad_to_multiple_of=self.pad_to_multiple_of,
            return_tensors="pt",
        )

        # Pad phoneme labels using the target processor
        with self.processor.as_target_processor():
            labels_batch = self.processor.pad(
                label_features,
                padding=self.padding,
                max_length=self.max_length_labels,
                pad_to_multiple_of=self.pad_to_multiple_of_labels,
                return_tensors="pt",
            )

        # Replace padding tokens in labels with -100 so they're ignored during loss computation.
        # NOTE: Ensure that your custom tokenizer has the correct pad_token_id for your 42-phoneme vocabulary.
        labels = labels_batch["input_ids"]
        labels[labels == self.processor.tokenizer.pad_token_id] = -100

        batch["labels"] = labels
        return batch

```

```python
from torch.utils.data import IterableDataset

class InputIdsWrapper(IterableDataset):
    def __init__(self, streaming_dataset):
        self.streaming_dataset = streaming_dataset

    def __iter__(self):
        for example in self.streaming_dataset:
            # If the example doesn’t already have an "input_ids" key,
            # add it by copying the "input_values"
            if "input_ids" not in example:
                example["input_ids"] = example["input_values"]
            yield example
```
